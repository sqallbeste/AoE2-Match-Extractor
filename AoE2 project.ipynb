```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
```

```python
url = "https://liquipedia.net/ageofempires/Warlords/3"

response = requests.get(url)
bool(response)
soup = BeautifulSoup(response.text, "html.parser")
print(response)
```

```python
#Cogemos todas las fichas de cada serie
ficha = soup.find_all("div", attrs = {"class":"brkts-popup brkts-match-info-popup"})

#Comenzamos a recopilar toda la información de cada una:
for i in range(len(ficha)-1):
    sesion = i
    print(sesion)
    
    nombres = ficha[sesion].find_all("a") 
    Player1 = nombres[0].text 
    Player2 = nombres[1].text 
    fecha = ficha[sesion].find("span", attrs = {"class":"timer-object"}).text
    print(Player1,Player2,fecha)
    
    #Excepción para cuando se suspende la serie por motivos varios (enfermedad,abandono, etc)
    if ficha[sesion].find("div", attrs={"class": "brkts-popup-header-opponent-score-left"}).text not in ["W", "FF"]: 
        
        #Sumamos el score de ambos jugadores para saber cuantas partidas debemos recorrer
        Score1 = int(ficha[sesion].find("div", attrs = {"class":"brkts-popup-header-opponent-score-left"}).text)
        Score2 = int(ficha[sesion].find("div", attrs = {"class":"brkts-popup-header-opponent-score-right"}).text)
        num_partidas = Score1+Score2

        #Analizamos cada partida individual
        match = ficha[sesion].find_all("div", attrs = {"class":"brkts-popup-body-element brkts-popup-body-game"})

        for i in range(num_partidas):
            Civ1= match[i].find_all("a")[0]["title"]
            Civ2= match[i].find_all("a")[2]["title"]
            map = match[i].find_all("a")[1]["title"]
            
            #Si el check es verde sabemos que ha ganado el Jugador 1, en caso contrario el Jugador 2
            if match[i].find_all("i")[0]["class"][2] == 'forest-green-text':
                print(Civ1,map,Civ2,Player1)
            else:
                print(Civ1,map,Civ2,Player2)

```
